```python
%matplotlib inline
from matplotlib import pyplot as plt 
import IPython.display as ipd
import numpy as np
from scipy.signal import lfilter
```

# Phasor

```python
# parameter setting 
amp = 1         # range [0.0, 1.0]
sr = 10000      # sampling rate, Hz / 1초에 만개 숫자 설정 
dur = 0.5       # in seconds / 얼만큼 긴 사인웨이브
freq = 100.0    # sine freqeuncy, Hz / 사이클(반복구간)이 1초에 몇개 있게 하냐 -> 소리 높낮이
```

```python
# generate time
t = np.arange(1, sr*dur+1)/sr
# 시간과 phase를 링크시킴.
# sr이 있어야 시간 만들수 있음. 
# 0에서 0.5초 까지 시간이 만들어진 것 
```

```python
# generate phase
theta = t*2*np.pi*freq  # 
# 시간 * 2파이 -> 한바퀴짜리 웨이브 
# 시간*2파이*100 -> 100바퀴짜리 웨이브 = Phase
# 0에서 2파이*100*0.5까지 radian값이 만들어진 것 
```

```python
# generate signal by cosine-phasor
s = np.sin(theta)   # theta -> in radian (radian값-각도값)
# 실질적으로 사인곡선이 만들어지는것  
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)   # subplot(234): 2*3의 창틀을 만듦.그중에서 4번째
ax.plot(t[0:1000],s[0:1000],'.')  # x축이 시간, y축이 사인웨이브 
# t의 샘플갯수는? 1000개 
ax.set_xlabel('theta (radian)')
ax.set_ylabel('signal')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)   # subplot(234): 2*3의 창틀을 만듦.그중에서 4번째
ax.plot(t,s,'.')  # x축이 시간, y축이 사인웨이브
# 1초에 10000개의 점 찍힘  
ax.set_xlabel('theta (radian)')
ax.set_ylabel('signal')
# 너무 빽빽하니까 1000개만 고르자 
```
